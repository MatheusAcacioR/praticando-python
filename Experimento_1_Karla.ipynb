```python
# Importando a biblioteca pandas e matplot

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Chamando o dataset oiginal e o dataset tratado

dt = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Extração 23Nov21 - Um ano/DadosInterpolados_23Nov21.csv")
dt
```

```python
# removendo as linhas 0 e 1 que sao linhas de texto e as linhas de dados outliers

dt = dt.drop([0, 1, 305917, 305918, 314845, 331464, 331465, 331466, 331467, 331468, 331469, 331470, 331471, 331472, 331473, 331474, 409704, 500411, 509200, 509201])
```

```python
# renomeando a primeira coluna para timestamp

dt.rename(columns={'Unnamed: 0': 'Timestamp'}, inplace=True)
```

```python
# resetando a indexação do dataframe

dt = dt.reset_index(drop = True)
```

```python
# convertendo as colunas numericas para float a coluna Timestamp para datetime

dt['B1001-EOF'] = dt['B1001-EOF'].astype(float)
dt['1TIC1001/SP'] = dt['1TIC1001/SP'].astype(float)
dt['1AY1010-1/PV'] = dt['1AY1010-1/PV'].astype(float)
dt['1TIC1001/PV'] = dt['1TIC1001/PV'].astype(float)
dt['Timestamp'] = pd.to_datetime(dt['Timestamp'], format='%m-%d-%Y %H:%M:%S')
```

```python
# renomeando as linhas da coluna B1001-EOF para Propano (O que for -9998.0) e Etano (O que for -9992.0)

dt.loc[dt["B1001-EOF"] == -9998.0, 'B1001-EOF'] = "Propano"
dt.loc[dt["B1001-EOF"] == -9992.0, 'B1001-EOF'] = "Etano"
dt.loc[dt["B1001-EOF"] == -9997.0, 'B1001-EOF'] = "Carga_piloto_apagado"
dt.loc[dt["B1001-EOF"] == -9996.0, 'B1001-EOF'] = "HSS"
dt.loc[dt["B1001-EOF"] == -9995.0, 'B1001-EOF'] = "Elev_resf"
dt.loc[dt["B1001-EOF"] == -9994.0, 'B1001-EOF'] = "NA"
dt.loc[dt["B1001-EOF"] == -9993.0, 'B1001-EOF'] = "Decoque"
dt.loc[dt["B1001-EOF"] == -9991.0, 'B1001-EOF'] = "Piloto_aceso"
dt.loc[dt["B1001-EOF"] == -9990.0, 'B1001-EOF'] = "Parado"
dt.loc[dt["B1001-EOF"] == -9989.0, 'B1001-EOF'] = "Shutdown"
```

## Temperatura

```python
# Limites: < 800 / > 855
# Linhas que estao fora do range de temperatura adequado

outOfRangeDegree = dt[(dt["1TIC1001/PV"] <= 799) | (dt["1TIC1001/PV"] >= 856)]
outOfRangeDegree
```

```python
# Os dias que houveram falha por temperatura

diasDeFalhaTemperatura = pd.DataFrame({
    "Dias de falhas de temperatura": outOfRangeDegree['Timestamp'].dt.date
}).drop_duplicates().reset_index(drop = True)

diasDeFalhaTemperatura
```

```python
# Grafico que mostra a porcentagem de falhas de temperatura dentro desses 1 ano de dataset

labels = 'Falha', 'Sem falha'
sizes = [len(diasDeFalhaTemperatura), 365]
figl, axl = plt.subplots()
c = ['#EB002A', '#DF0CF6']
axl.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False, startangle=90, colors=c)
axl.axis('equal')
axl.set_title("Porcentagem de falhas pela temperatura em 1 ano")
plt.show()
```

```python
# Quantas vezes (em minutos) teve falha por temperatura em cada dia que houve falha

agrupamento = pd.DataFrame({
    "Dias": outOfRangeDegree['Timestamp'].dt.date,
    "1TIC1001/PV": outOfRangeDegree['1TIC1001/PV']
}).reset_index(drop = True).groupby(["Dias"]).size()

agrupamento
```

```python
# Linhas de propano fora da faixa de temperatura

outOfRangeDegreePropano = outOfRangeDegree[outOfRangeDegree["B1001-EOF"] == 'Propano']

outOfRangeDegreePropano
```

```python
# Dias que houveram falha de temperatura e que eram Propano

diasDeFalhaPropanoTemperatura = pd.DataFrame({
    "Dias de falhas de temperatura": outOfRangeDegree['Timestamp'].dt.date
}).reset_index(drop = True)

diasDeFalhaPropanoTemperatura
```

```python
dias = diasDeFalhaPropanoTemperatura["Dias de falhas de temperatura"]
dias_sum = dias.value_counts()
dias_sum = dias_sum.sort_index()

x = dias_sum.index
y = dias_sum

plt.figure(figsize=(8, 6))

plt.bar(x, y)

plt.title('Quantidade de falhas de Propano nos dias')
plt.xlabel('Dias')
plt.xticks(rotation=90)
plt.ylabel('Falhas')
plt.xticks(x) # obriga a mostrar todos os números no eixo x
```

```python
# Limites: < 60 / > 75
# Linhas de Etano que estao fora do range de sua porcentagem
outOfRangeEtan = dt[((dt["1AY1010-1/PV"] <= 59) | (dt["1AY1010-1/PV"] >= 76)) & (dt["B1001-EOF"] == 'Etano')]
outOfRangeEtan
```

```python
# Grafico que mostra a porcentagem de falhas de etano dentro desses 1 ano de dataset

labels = 'Etano', 'Sem falha'
sizes = [len(quantidadePorDiaEtano), 365]
figl, axl = plt.subplots()
c = ['#EB002A', '#DF0CF6']
axl.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False, startangle=90, colors=c)
axl.axis('equal')
axl.set_title("Porcentagem de falhas pelo Etano em 1 ano")
plt.show()
```
